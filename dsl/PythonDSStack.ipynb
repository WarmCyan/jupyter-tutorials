### Classes

```python
class life:
    def __init__(self, name, age, dead_inside):
        self.name = name
        self.age = age
        self.dead_inside = dead_inside
    
    def death_status(self):
        print("\n", self.age, self.dead_inside) # string concatenation v0
        if self.dead_inside:
            print("Poor {0} - {1}".format(self.name, "he's dead inside")) # string concatenation v1
        else:
            print(self.name + " is happy and healthy!") # string concatenation v2
```

```python
nathan = life("Nathan", 22, True)
nathan.info()

nathan_alternate_universe = life("Nathan_1", 22, False)
nathan_alternate_universe.info()
```

# Matplotlib

Documentation: https://matplotlib.org/

```python
import matplotlib.pyplot as plt
%matplotlib notebook

from sklearn import datasets
```

```python
iris = datasets.load_iris()
iris.keys()
```

```python
df_iris = pd.DataFrame(iris.data, columns=iris.feature_names)
df_iris.head()
```

```python
plt.figure()

plt.hist(df_iris["sepal length (cm)"], bins=25)
plt.show()
```

```python
plt.figure()
plt.scatter(df_iris["sepal length (cm)"], df_iris["petal length (cm)"])
plt.show()
```

```python
plt.figure()
plt.scatter(df_iris["sepal length (cm)"], df_iris["petal length (cm)"], df_iris["petal width (cm)"]*10)
plt.show()
```

```python
iris.target
```

```python
df_iris_all = df_iris.copy()
df_iris_all['target'] = iris.target
df_iris_all.head()
```

```python
fig, axes = plt.subplots()
axes.clear()

axes.scatter(x=df_iris["sepal length (cm)"], y=df_iris["petal length (cm)"], s=df_iris["petal width (cm)"]*10, c=df_iris_all["target"])
axes.set_xlabel("sepal length")
axes.set_ylabel("petal length")
axes.grid(True)
```

# Scikit-Learn

Documentation: https://scikit-learn.org/stable/documentation.html

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=3)
model.fit(iris.data)
labels = model.labels_
labels
```

```python
df_iris_predicted = df_iris.copy()
df_iris_predicted["predicted"] = labels
df_iris_predicted.head()
```

```python
fig = plt.figure()
axes1 = fig.add_subplot(1,2,1)
axes2 = fig.add_subplot(1,2,2)

axes1.scatter(x=df_iris["sepal length (cm)"], y=df_iris["petal length (cm)"], s=df_iris["petal width (cm)"]*10, c=df_iris_all["target"])
axes2.scatter(x=df_iris["sepal length (cm)"], y=df_iris["petal length (cm)"], s=df_iris["petal width (cm)"]*10, c=df_iris_predicted["predicted"])

axes1.set_title("Actual")
axes2.set_title("Predicted")
```
